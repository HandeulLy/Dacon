# 데이터 로드

```python
import pandas as pd
```

```python
train = pd.read_csv(path + 'train.csv')
test = pd.read_csv(path + 'test_x.csv')
submission = pd.read_csv(path + 'sample_submission.csv')
```

```python
print(train.shape)
print(test.shape)
print(submission.shape)
```

# 환경 구축

```python
# 'voted' 컬럼이 예측 대상이므로 target 인자에 명시
# 'voted' column is the target variable
clf = setup(data = train, target = 'voted')
```

# 모델 학습 비교

```python
best_3 = compare_models(sort = 'AUC', n_select = 3)
```

# 모델 앙상블

```python
blended = blend_models(estimator_list = best_3, fold = 3, method = 'soft')
```

# 예측

```python
pred_holdout = predict_model(blended)
```

# 재학습

```python
final_model = finalize_model(blended)
```

# Test Set 예측

```python
predictions = predict_model(final_model, data = test)
predictions
```

# 제출 파일 생성

```python
submission['voted'] = predictions['Score']
submission.to_csv('HDLY_201112_02.csv', index = False)
```
